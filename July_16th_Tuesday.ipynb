```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("fastfood.csv")

#print(data["restaurant"].unique())

calories = pd.pivot_table(data, values='calories', index = "restaurant", aggfunc="mean")

sort_calories = calories.sort_values(by="calories", ascending = True)

print(sort_calories)

plt.bar(sort_calories.index, sort_calories["calories"])
plt.xticks(rotation=90)
plt.savefig("caloriesperrestaurant.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("fastfood.csv")

#print(data["restaurant"].unique())

calories = pd.pivot_table(data, values='sat_fat', index = "trans_fat", aggfunc="mean")

sort_calories = calories.sort_values(by="sat_fat", ascending = True)

print(sort_calories)

plt.scatter(sort_calories.index, sort_calories["sat_fat"])
plt.xticks(rotation=90)
plt.savefig("satandtransfat.png")
plt.show()
```

```python
plt.figure(figsize=(20,6))
for restaurant in data['restaurant'].unique():
  subset = data[data['restaurant'] == restaurant]
  print(restaurant, subset["trans_fat"].max())
  print(subset.loc[subset["trans_fat"].idxmax()])
  plt.scatter(subset['item'], subset['trans_fat'], label=restaurant, alpha=0.7)

plt.legend()
plt.xticks()
plt.savefig("transrestauranthighest.png")
plt.show()
```

```python
boxplot = data.boxplot(column='calories', by = 'restaurant', grid=False)
plt.xticks(rotation=90)
plt.savefig("caloriesandrestaurant.png")
plt.show()
```

```python
boxplot = data.boxplot(column='trans_fat', by = 'restaurant', grid=False)
plt.xticks(rotation=90)
plt.savefig("restaurantpertransfat.png")
plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.scatter(data['calories'], data['trans_fat'], color = 'blue')
plt.xlabel("calories")
plt.ylabel("trans_fat")
plt.savefig("calories_transfat_relatioshipscatter.png")
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(data=data, x='calories', y='trans_fat', alpha = 0.7, s=100)
sns.regplot(data=data, x='calories', y='trans_fat', scatter=False, color='red')
plt.title('Relationship between Calories and Trans fat')
plt.xlabel('Calories')
plt.ylabel('Trans fat(g)')
plt.savefig("Calories_Transfat_relationship.png")
plt.show()
```

```python
sns.scatterplot(data=data, x='protein', y='trans_fat', alpha = 0.7, s=100)
sns.regplot(data=data, x='protein', y='trans_fat', scatter=False, color='red')
plt.title('Relationship between protein and Trans fat')
plt.xlabel('protein')
plt.ylabel('Trans fat(g)')
plt.savefig("protein_tranfat_relationship.png")
plt.show()
```

```python

corr = data[["calories","cal_fat","total_fat","trans_fat","protein"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Niutritional Factors")
plt.savefig("heatmap.png")
plt.show()
```
